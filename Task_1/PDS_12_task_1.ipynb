**Calculating Plant Growth as a Function of Environmental Factors**

While the dataset directly provides plant growth measurements, understanding plant growth often involves considering its relationship with environmental factors. A crucial aspect of this analysis is examining plant growth as a function of temperature and light intensity. We can assume that plant growth is influenced by these factors. To investigate this relationship, we need to analyze how changes in temperature and light intensity correspond to variations in plant growth. This exploration will provide insights into the impact of these environmental conditions on plant development.

```python
#make a plot of 33*9 size
fig_h = plt.figure(figsize=(33,9))
#add the first subplot dividing the plot into 1 row and two columns and add subplot to first column
fig1 = fig_h.add_subplot(1,2,1)
#make histogram to show the distribution of plant growth under varying temperature
sns.histplot(data, y= "Plant_Growth (cm)", x="Temperature (°C)",hue="Plant_Growth (cm)", element="poly")
#add title and axis labels and legends
fig1.set_title("Plant Growth Distribution Under Varying Temperature Conditions")
#add title and axis labels and legends
fig1.set_xlabel("Temperature (°C)")
fig1.set_ylabel("Plant Growth (cm)")
legend = fig1.get_legend()
legend.set_bbox_to_anchor((1, .6))

#add the first subplot dividing the plot into 1 row and two columns and add subplot to second column
fig2 = fig_h.add_subplot(1,2,2)
#make histogram to show the distribution of plant growth under varying light intensity
sns.histplot(data, y = "Plant_Growth (cm)", x= "Light_Intensity (lux)", hue="Plant_Growth (cm)", element="poly")
fig2.set_title("Plant Growth Distribution Under Varying Temperature Conditions")
fig2.set_xlabel("Temperature (°C)")
fig2.set_ylabel("Plant Growth (cm)")
legend2 = fig2.get_legend()
legend2.set_bbox_to_anchor((1, .6))

#adjust spacing between two subplots
fig_h.subplots_adjust(wspace=0.1)

#show the histograms
plt.show()
```

The two histogram show the distribution of plant growth under varying temperature and light intensity.

Temperature: Plant growth seems to peak at a certain temperature range, following a bell curve pattern. This suggests an optimal temperature for growth.


Light Intensity: The relationship between light intensity and growth appears more complex and less clear-cut. It could be affected by factors beyond just intensity.

```python
# #make a plot of 33*9 size with 1 row, 2 columns
fig, axes = plt.subplots(2, 1, figsize=(16, 19))

# Box plot for Temperature vs. Plant Growth, axes[0] refers to the first subplot
sns.boxplot(x="Temperature (°C)", y="Plant_Growth (cm)", data=data, ax=axes[0])
axes[0].set_title("Plant Growth Distribution Under Varying Temperature Conditions")

# Box plot for Light Intensity vs. Plant Growth, axes[1] refers to the second subplot
sns.boxplot(x="Light_Intensity (lux)", y="Plant_Growth (cm)", data=data, ax=axes[1])
axes[1].set_title("Plant Growth Distribution Under Varying Light Intensity Conditions")

# Adjust the spacing between subplots
fig.subplots_adjust(hspace=0.15)

# Display the plots
plt.show()
```

The box plots compare plant growth under varying temperature and light intensity.

Temperature: Plant growth generally increases with higher temperatures, with a clearer pattern and greater variability at higher temperatures.

Light Intensity: The relationship between light intensity and plant growth is less clear and shows less variability across intensity categories.

```python
#make a plot of 33*9 size
fig_s = plt.figure(figsize=(33, 9))

# Create the first subplot (1 row, 2 columns, first subplot)
fig1 = fig_s.add_subplot(1, 2, 1)

# Create a scatter plot of Plant Growth vs. Temperature
sns.scatterplot(x="Temperature (°C)", y="Plant_Growth (cm)", data=data, color="red")

#set plot title and axis labels
fig1.set_title("Plant Growth vs Temperature")
fig1.set_xlabel("Temperature (°C)")
fig1.set_ylabel("Plant Growth (cm)")


# Create the second subplot (1 row, 2 columns, second subplot)
fig2 = fig_s.add_subplot(1, 2, 2)

# Create a scatter plot of Plant Growth vs. Light Intensity
sns.scatterplot(y="Plant_Growth (cm)", x="Light_Intensity (lux)", data=data, color="blue")

#set plot title and axis labels
fig2.set_title("Plant Growth vs Light Intensity")
fig2.set_xlabel("Light Intensity (lux)")
fig2.set_ylabel("Plant Growth (cm)")

# Adjust spacing between subplots
fig_s.subplots_adjust(wspace=0.1)

# Display the plots
plt.show()
```

The scattrplots shows relationship of plant growth with varying temperature and light intensity.


Temperature: Shows a clear positive correlation with plant growth. Higher temperatures generally lead to increased growth.


Light Intensity: Shows a weaker, possibly more complex relationship with plant growth, with more variability and scatter in the data.

```python
# Calculate the correlation matrix
correlation_matrix = data[["Temperature (°C)", "Light_Intensity (lux)", "Plant_Growth (cm)"]].corr()


# Display the correlation matrix (optional - you can remove this line if you don't need to see the raw matrix)
correlation_matrix

# Set the figure size to 8*6
plt.figure(figsize=(8, 6))

# Create a heatmap to display the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='prism_r', vmin=-1, vmax=1)
# Set the title of the heatmap
plt.title('Correlation Matrix')

# Display the heatmap
plt.show()
```

The heatmap visually represents the correlation between 'Temperature (°C)', 'Light_Intensity (lux)', and 'Plant_Growth (cm)'.

Temperature & Plant Growth: Shows a strong positive correlation (the darker red color indicates a stronger relationship), implying that as temperature increases, plant growth tends to increase as well.


Light Intensity & Plant Growth: Shows a weaker positive correlation (lighter color) compared to temperature, suggesting a less direct or more complex relationship.


Temperature & Light Intensity: Shows a very weak or no correlation, indicating these two factors are likely independent of each other in this dataset.

```python
# Find the strongest correlation with 'Plant_Growth (cm)'
# by excluding self-correlation and taking the absolute maximum
strongest_corr = correlation_matrix['Plant_Growth (cm)'].drop('Plant_Growth (cm)').abs().idxmax()
# Print the result
print(f"The Strongest correlation with plant growth is: {strongest_corr}")
```

```python
#make a plot of 25*9 size
fig_l = plt.figure(figsize=(25,9))
#add the first subplot dividing the plot into 1 row and two columns and add subplot to first column
fig1 = fig_l.add_subplot(1,2,1)

# Create a scatter plot with a regression line to visualize the relationship between Temperature and Plant Growth
sns.regplot(x="Temperature (°C)", y='Plant_Growth (cm)', data=data, ax=fig1)
fig1.set_title('Linear Regression: Temperature vs. Plant Growth')


#add the first subplot dividing the plot into 1 row and two columns and add subplot to second column
fig2 = fig_l.add_subplot(1,2,2)

# Create a scatter plot with a regression line to visualize the relationship between Light Intensity and Plant Growth
sns.regplot(x="Light_Intensity (lux)", y='Plant_Growth (cm)', data=data, ax=fig2)
fig2.set_title('Linear Regression: Light Intensity vs. Plant Growth')

# Adjust spacing between subplots
fig_l.subplots_adjust(wspace=0.1)
#display the plot
plt.show()
```

Linear Regression: Temperature vs. Plant Growth

Shows a clear positive linear relationship between temperature and plant growth. As temperature increases, plant growth tends to increase as well. The regression line reinforces this positive trend. So, Temperature is a significant factor positively influencing plant growth within the observed range, and this relationship is primarily linear.


Linear Regression: Light Intensity vs. Plant Growth

Shows a weaker, less clear relationship between light intensity and plant growth. There's more scatter around the regression line, suggesting other factors or interactions are likely influencing plant growth alongside light intensity. While light intensity might play a role, the relationship with plant growth is more complex and could be non-linear or influenced by other variables not captured in this visualization.

```python
# Convert the 'Date' column to datetime objects
data['Date'] = pd.to_datetime(data['Date'])

# Set the 'Date' column as the index of the DataFrame
data.set_index('Date', inplace=True)
```

The plot indicates that plant growth has generally increased over the time period captured in the data. While there are periods of slower growth or even slight declines, the overall trajectory suggests a positive growth pattern.

```python
# Normalize data for comparison
data_normalized = data[['Plant_Growth (cm)', 'Temperature (°C)', 'Light_Intensity (lux)']].apply(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
data_normalized
# Plot the normalized data
plt.figure(figsize=(12, 6))

plt.bar(data.index, data_normalized['Plant_Growth (cm)'], color='green', alpha=0.6, label='Plant Growth (cm)', width=1)
plt.plot(data.index, data_normalized['Temperature (°C)'], label='Temperature (°C)', color='red', marker='o')
plt.plot(data.index, data_normalized['Light_Intensity (lux)'], label='Light Intensity (lux)', color='blue', marker='o')

# Titles and labels

# Titles and labels
plt.title('Normalized Plant Growth, Temperature, and Light Intensity Over Time', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Normalized Values', fontsize=12)

# Legend and grid
plt.legend()
plt.grid(alpha=0.5)

# Show the plot
plt.show()
```

The figure shows normalized plant growth, temperature, and light intensity over time.

Plant growth (green bars) generally increases over time, but there are fluctuations.
Temperature (red line) and light intensity (blue line) also vary over time and may influence growth patterns.
The normalization allows for easier comparison of the three variables on the same scale.
